```python
def convert_to_bytes(inp):
    if inp is not None:
        print("UTF-8 Encoding: ", inp.encode('utf-8'))
        print("UTF-16 Encoding: ", inp.encode('utf-16'))
        print("ASCII Encoding: ", inp.encode('ascii'))
    else:
        print("Input is None")

convert_to_bytes("Python Programming!")
```

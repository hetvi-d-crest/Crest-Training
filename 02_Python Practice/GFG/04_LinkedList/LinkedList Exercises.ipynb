**Print Linked List**

```python
class Node:
    def __init__(self, val = None):
        self.val = val
        self.next = None

class LinkedList:
    def __init__(self, head = None):
        self.head = head

    def __str__(self):
        res = ''

        curr = self.head

        while curr:
            res += str(curr.val) + ', '
            curr = curr.next

        res = res.strip(', ')

        if len(res):
            return '[' + res + ']'
        else:
            return '[]'

if __name__ == "__main__":
    l1 = LinkedList()

    n1 = Node(10)
    n2 = Node(20)
    n3 = Node(30)

    l1.head = n1
    n1.next = n2
    n2.next = n3

    print(l1)
    
```

**Stack using Doubly Linked List**

```python
class Node:
    def __init__(self, val = None, next = None, prev = None):
        self.val = val
        self.next = next
        self.prev = prev

class Stack:
    def __init__(self):
        self.head = None

    def push(self, data):
        if self.head is None:
            self.head = Node(data)
        else:
            new_node = Node(data)
            self.head.prev = new_node
            new_node.next = self.head
            new_node.prev = None
            self.head = new_node

    def pop(self):
        if self.head is None:
            return None
        elif self.head.next is None:
            temp = self.head.val
            self.head = None
            return temp
        else:
            temp = self.head.val
            self.head = self.head.next
            self.head.prev = None
            return temp
            

    def top(self):
        return self.head.val

    def size(self):
        temp = self.head
        count = 0
        while temp is not None:
            count +=1
            temp = temp.next
        return count

    def isEmpty(self):
        return self.head is None

    def printstack(self):
        print("Stack elements are: ")
        temp = self.head
        while temp is not None:
            print(temp.val, end = " ->")
            temp = temp.next
        print("None")


if __name__ == "__main__":
    stack = Stack()

    print("Stack operations using Doubly LinkedList")
    stack.push(10)
    stack.push(50)
    stack.push(23)
    stack.push(21)
    stack.push(21)


    stack.printstack()
    print("Top Element is: ", stack.top())
    print("Size of the stack is", stack.size())

    stack.pop()
    stack.pop()

    stack.printstack()
    print("\nStack is empty:", stack.isEmpty())
```

**Queue using Linked List**

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.front = None
        self.rear = None

    def enqueue(self, data):
        new_node = Node(data)
        if self.rear == None:
            self.rear = self.front = new_node
        else:
            self.rear.next = new_node
            self.rear = self.rear.next

    def dequeue(self):
        if self.front is None:
            print("Empty Queue")
            return None
        else:
            temp = self.front
            self.front = self.front.next

        if self.front is None:
            self.rear = None
        return temp.data
        
    def peek(self):
        if self.front is None:
            print("Empty Queue")
            return None
        return self.front.data

    def isEmpty(self):
        return self.front is None

    def size(self):
        cnt = 0
        curr = self.front
        while curr is not None:
            cnt += 1
            curr = curr.next
        return cnt

    def printqueue(self):
        print("queue elements are: ")
        curr = self.front
        while curr:
            print(curr.data, end = " -> ")
            curr = curr.next
        print("None")

    
            
if __name__ == "__main__":
    q = Queue()
    print("Queue operations using Linked List")
    q.enqueue(10)
    q.enqueue(20)
    q.enqueue(30)
    q.enqueue(40)

    q.printqueue()
    print("Front element is:", q.peek())
    print("Queue size is:", q.size())

    q.dequeue()
    q.dequeue()

    q.printqueue()
    print("Queue is empty:", q.isEmpty())        
```

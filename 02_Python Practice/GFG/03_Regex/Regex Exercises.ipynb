```python
import re
```

```python
s = 'asdasd123asmdasdk34234kfdsd34sdfk5'
n = re.findall(r'\d+',s)
print(n)
print(" ".join(n))
```

```python
s = 'asdsab@!@234'
pattern = r'^[a-z]+[!@#$%]+[0-9]+$'
print(bool(re.fullmatch(pattern, s)))
```

```python
string = 'barfoobarfoobarfoobarfoobarfoo'
pattern = 'foobarfoo'

match = [m.start() for m in re.finditer(f'(?={pattern})', string)]
print(match)
```

```python
def check(string, pattern):
    if re.search(pattern, string):
        print("Valid string")
    else:
        print("Invalid string")

pattern = re.compile('^[1234]+$')
check('214', pattern)
check('215', pattern)
```

```python
filenames = ["gfg.html", "geeks.xml", "computer.txt", "geeksforgeeks.jpg"]

for file in filenames:
    match = re.search("\.xml$", file)

    if match:
        print("the file ending with .xml is: ", file)
```

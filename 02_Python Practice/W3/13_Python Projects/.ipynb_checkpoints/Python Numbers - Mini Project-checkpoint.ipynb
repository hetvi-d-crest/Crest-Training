```python
from math import pi

def get_pi(n):
    return round(pi, n)

inp = input("Enter the number of digits you want after the decimal for Pi")
try:
    user_input = int(inp)
    print(get_pi(user_input))
except:
    print("You did not enter an integer")
```

```python
get_pi(2)
```

```python
import random

random_num = random.randrange(0,100)
guess = False

while not guess:
    try:
        inp = int(input("Please enter a number between 0 and 100"))
    except ValueError:
        print("This is not a valid integer input. Try again!")
        continue
    if inp < random_num:
        print("The number you entered is less then the actual number. Try again")
    elif inp > random_num:
        print("The number you entered is more than the actual number. Try again")
    else:
        print("You guessed it right!")
        guess = True
```

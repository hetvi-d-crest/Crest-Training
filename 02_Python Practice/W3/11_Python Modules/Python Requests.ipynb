```python
import requests

print("timeout : 0.001")
try:
    res = requests.get('https://github.com/', timeout = 0.001)
    print(res.text)
except requests.exceptions.RequestException as e:
    print("An error ccured: ", e)

print("\ntimeout = 1.0")    
try:
    r = requests.get('https://github.com/', timeout = 1.0)
    print("Connected....!")
except requests.exceptions.RequestException as e:
    print(e)
    
```

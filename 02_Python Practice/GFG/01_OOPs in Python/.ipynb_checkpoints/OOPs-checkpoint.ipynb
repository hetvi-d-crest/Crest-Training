**Python program to build flashcard using class**

```python
class Flashcard:
    def __init__(self, word, meaning):
        self.word = word
        self.meaning = meaning

    def __str__(self):
        return self.word + '(' + self.meaning + ')'

f = []

while True:
    word = input("Please enter the word: ")
    meaning = input("Please enter its meaning")
    f.append(Flashcard(word, meaning))
    valid = input("Press 'Y' if you want to enter more cards")

    if valid.lower() != 'y':
        break
        
print("The flash cards are:")
for cards in f:
    print(cards)
```

```python
import unittest

def check_sorted(lst):
    return all(lst[i] <= lst[i+1] for i in range(len(lst) - 1))

class TestSorting(unittest.TestCase):
    def is_True(self):
        lst = [1, 2, 3, 4, 5, 6, 7]
        print("Sorted list: ",lst)

        self.assertTrue(check_sorted(lst), "The list is not sorted in ascending order")

    def is_False(self):
        lst = [5, 7, 2, 8, 1, 9]

        print("Unsorted list: ", lst)
        self.assertFalse(check_sorted(lst), "The list is sorted in ascending order")

if __name__ == '__main__':
     unittest.main(argv=['first-arg-is-ignored'], exit=False)
    
```

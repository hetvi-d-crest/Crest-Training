```python
import re

pattern = r'\d{3}-\d{3}-\d{4}'

def search(file):
    f = open(file, 'r')
    content = f.read()

    if re.search(pattern, content):
        return re.search(pattern, content)
    else:
        return ''
        
```

```python
import os
results = []

for folders, subfolders, files in os.walk(os.getcwd()+'\\extracted_content'):
    for f in files:
        full_path = folders +'/'+f
        results.append(search(full_path))

print(results)
```

```python
for r in results:
    if r != '':
        print(r.group())
```

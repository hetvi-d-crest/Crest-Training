```python
def main():
    try:
        frozenset_x = frozenset([1, 2, 3, 4, 5])
        frozenset_y = frozenset([0, 1, 3, 7, 8, 10])   
        print("Original frozensets: ")
        print("Frozenset x: ",frozenset_x)
        print("Frozenset y: ",frozenset_y)
        print("Union: ")
        print(frozenset_x.union(frozenset_y))
        print("Intersection: ")
        print(frozenset_x.intersection(frozenset_y))
        print("difference of (frozenset_x-frozenset_y): ")
        print(frozenset_x.difference(frozenset_y))
        print("difference of (frozenset_y-frozenset_x): ")
        print(frozenset_y.difference(frozenset_x))
    except Exception as e:
        print("An exception occured!", e)
        
if __name__ == '__main__':
    main()
```

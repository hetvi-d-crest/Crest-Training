```python
def check(arr):
    fives = 3
    nineteens = 2

    for n in arr:
        if nineteens <0:
            return False
        else:
            if n == 19:
                nineteens -= 1
            if n == 5:
                fives -= 1


    if fives > 0:
        return False

    return True

check([19, 19, 15, 5, 3, 5, 5, 2])            
```

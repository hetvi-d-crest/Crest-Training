**Linear Search**

**Iterative Approach**

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i

    return -1

arr = [10, 23, 70, 11, 15]
target = 00

result = linear_search(arr, target)

if result != 1:
    print(f"Element found at index: {result}")
else:
    print("Element not found")
```

**Recursive Approach**

```python
def recursive_search(arr, target, i =0):
    if len(arr) == i:
        return -1

    if arr[i] == target:
        return i

    return recursive_search(arr, target, i+1)

result = recursive_search(arr, target)


if result != -1:
    print(f"Element {target} found at index {result}.")
else:
    print(f"Element {target} not found in the list.")
```

**Binary Search**

```python
def binary_search(arr, target):
    low, high = 0, len(arr)-1

    while low < high:
        mid = (low+high)//2
        if arr[mid] > target:
            high -=1
        elif arr[mid] < target:
            low += 1

        else:
            return mid
    return -1

arr = [ 2, 3, 4, 10, 40 ]
target = 10

result = binary_search(arr, target)

if result != 1:
    print(f"Element found at index: {result}")
else:
    print("Element not found")
```

```python
Max = 256

def compare(arr1, arr2):
    for i in range(Max):
        if arr1[i] != arr2[i]:
            return False
        else:
            return True

def search(pat, txt):
    M = len(pat)
    N = len(txt)

    cntp = [0]*Max
    cnttw = [0]*Max

    for i in range(M):
        (cntp[ord(pat[i])]) += 1
        (cnttw[ord(txt[i-M])]) -= 1

    if compare(cntp, cnttw):
        print("Found at index", N-M)

txt = "BACDGABCDA"
pat = "ABCD"       
search(pat, txt)
```

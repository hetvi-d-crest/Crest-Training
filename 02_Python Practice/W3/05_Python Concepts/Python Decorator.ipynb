```python
def decorator(func):
    def wrap(*args, **kwargs):
        print(f"Calling {func.__name__} with args: {args}, kwargs: {kwargs}")

        result = func(*args, **kwargs)

        print(f'{func.__name__} returned : {result}')

        return result
    return wrap
```

```python
@decorator
def multiply(x, y):
    return x*y

multiply(10, 5)
```

```python
def gensquares(N):
    for i in range(N):
        yield i*i

for i in gensquares(4):
    print(i)
```

```python
import random

def genrandom(low, high, n):
    for i in range(n):
        yield random.randint(low, high)

for i in genrandom(1, 10, 4):
    print(i)
    
```

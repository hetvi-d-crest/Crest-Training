```python
class MaxFinder:
    def __init__(self, numbers):
        self.numbers = numbers

    def findMax(self):
        currMax = float('-inf')

        for n in self.numbers:
            if n > currMax:
                currMax = n

        return currMax

obj = MaxFinder([1,2,3,4,5,6,1,100, 0, 4])
obj.findMax()
    
```

**Inheritance**

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speaks(self):
        print(f'{self.name} makes a sound.')

class Dog(Animal):
    def speaks(self):
        print(f'{self.name} barks')

class Lion(Animal):
    def speaks(self):
        print(f'{self.name} roars')

a= Animal("Animals")
d = Dog("Dog")
l = Lion("Lion")

a.speaks()
d.speaks()
l.speaks()
```

**Polymorphism**

```python
class Animal:
    def speak(self):
        print("Animal class:")

class Dog(Animal):
    def speak(self):
        print("Dog speaks")

class Cat(Animal):
    def speak(self):
        print("Cat speaks")


animals = [Animal(), Dog(), Cat()]

for a in animals:
    a.speak()
```

**Types of variable**

```python
global_var = "GLOBAL"

class Parent:
    class_var = "CLASS VAR"

    def __init__(self, name):
        self.name = name

    def display(self):
        print("The global var is ", global_var)
        print("The class var is : ", Parent.class_var)
        print("The class Parent's name is : ", self.name)

p = Parent("Sudha")
p.display()
```

**Types of Methods**

```python
class Demo:
    # Class variable
    class_name = "DemoClass"

    def __init__(self, value):
        # Instance variable
        self.value = value

    # Instance Method
    def instance_method(self):
        print("Instance Method Called")
        print(f"Value: {self.value}")

    # Class Method
    @classmethod
    def class_method(cls):
        print("Class Method Called")
        print(f"Class Name: {cls.class_name}")

    # Static Method
    @staticmethod
    def static_method():
        print("Static Method Called")
        print("This method does not use instance or class data")
obj = Demo(42)

obj.instance_method()
print()
obj.class_method()
print()
obj.static_method()
```

**Operator Overloading**

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

    def __str__(self):
        return f"({self.x}, {self.y})"

p1 = Point(1, 2)
p2 = Point(3, 4)

result = p1 + p2

print("Result is", result)
```

**Method Overloading**

```python
class Calc:
    def calcSum(self, a, b = 0, c = 0):
        print("The sum is" , a+b+c)

c= Calc()
c1 = Calc()
c2 = Calc()

c.calcSum(1)
c1.calcSum(1,2)
c2.calcSum(1,2,3)
```

**Method Overriding**

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

a = Animal()
d = Dog()

a.speak() 
d.speak()
```

**Inner Class**

```python
class Outer:
    def __init__(self):
        self.name = "Outer Class"

    def show(self):
        print("This is the outer class")

    class Inner:
        def __init__(self):
            self.name = "Inner Class"

        def display(self):
            print("This is the inner class")

outer = Outer()
outer.show()

inner = Outer.Inner()
inner.display()
```

**Random**
1. Random Value Generation

Write a Python program to generate a random color hex, a random alphabetical string, random value between two integers (inclusive) and a random multiple of 7 between 0 and 70. Use random.randint()

```python
import random
import string

print("Random Color hex")
print("#{:06x}".format(random.randint(0, 0xFFFFFF)))
print()
print("Random alphabetical string")
max_length = 255
s = ""
for i in range(random.randint(1, max_length)):
    s += random.choice(string.ascii_letters)
print(s)
print()
print("Random value between two integers, inclusive")
print(random.randint(1, 10))
print()
print("Random multiple of 7 between 0 and 70")
print(random.randint(0, 10)*7)
```

1. User-Defined Function Checker

Write a Python program to check if a function is a user-defined function or not. Use types.FunctionType, types.LambdaType()

```python
import types

def func():
    return 1

print(isinstance(func, types.FunctionType))
print(isinstance(func, types.LambdaType))
print(isinstance(lambda x : x+1, types.FunctionType))
print(isinstance(lambda x: x+1, types.LambdaType))
print(isinstance(max, types.FunctionType))
print(isinstance(max, types.LambdaType))

```

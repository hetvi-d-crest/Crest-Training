```python
def upperConvert(string):
    if not string:
        return None
    return string.upper()

def main():
    try:
        inp = input("Please enter a string to convert in upper case : ")
        res = upperConvert(inp)

        if res is None:
            print("Input is empty")
        else:
            print(res)
    except Exception as e:
        print("Exeption occured! : ", e)

if __name__ == '__main__':
    main()
            
```

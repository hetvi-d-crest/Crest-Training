```python
def getSum(arr):
    if len(arr) == 1:
        return arr[0]
    else:
        return arr[0] + getSum(arr[1:])
```

```python
list1 = [1,2,3,4,5]
getSum(list1)
```

```python
def conversion(num, base):
    convertString = '0123456789ABCDEF'

    if num < base:
        return convertString[num]
    else:
        return conversion(num // base, base) + convertString[num%base]
    
```

```python
print(conversion(2835, 16))
```

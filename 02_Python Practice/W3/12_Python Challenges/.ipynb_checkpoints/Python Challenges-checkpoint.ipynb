```python
def check(n):
    return n >= 0 and (n & (n-1)) == 0

check(4)
```

```python
check(16)
```

```python
check(100)
```

```python
def power_of_three(n):
    while n%3 == 0:
        n /= 3
    return n == 1 

power_of_three(9)
```

```python
power_of_three(100)
```

```python
def is_perfect_square(n):
    x = n // 2
    y = set([x])
    while x * x != n:
        x = (x + (n // x)) // 2
        if x in y: return False
        y.add(x)
    return True

print(is_perfect_square(8))
```

```python
print(is_perfect_square(16))
```

```python
print(is_perfect_square(25))
```

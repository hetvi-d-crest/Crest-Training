```python
from itertools import accumulate
import operator

def running_product(it):
    return accumulate(it, operator.mul)

list1 = [1,2,3,4,5,6]
print("The running product of a list:")
res = running_product(list1)
for i in res:
    print(i)

tuple1 = (1,2,3,4,5)
print("The running product of a tuple:")
ans = running_product(tuple1)
for i in ans:
    print(i)
```
